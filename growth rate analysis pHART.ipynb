```python
pHART_growth_data = pd.read_excel('round 2 compiled surface areas.xlsx','Sheet1') 
pHART_growth_data[:10]
```

```python
# There are some errors in the data. Sometimes the areas get smaller after after more days (areas should only increase or stay the same)
# Usually this is due to measurement error between closely-sized images. I will deal with this afterwards.
# However, sometimes the measurements are due to more grievious scaling errors. 
# I want to flag these up so that I can go back and check my measurements. I will use a cut-off of 0.1 mm^2 difference

pHART_growth_data ['Area_Change'] = pHART_growth_data.groupby('ID')['Area'].transform(pd.Series.diff) 
pHART_growth_data[pHART_growth_data ['Area_Change']<-0.1 ]

# there's only one, and I deem this to be small enough to call a 'smaller' measurement error
```

```python
# import the line names from a separate file
lines = pd.read_excel('round 2 compiled surface areas.xlsx','Lines') 
lines.set_index('Number', inplace=True)
lines [:10]
```

```python
pHART_growth_data['Line'] = lines.ix[pHART_growth_data['ID']].values
pHART_growth_data [:10]
```

```python
# Now I want to deal with the smaller measurement errors.  
# If the growth area decreases with increasing days, I will assign the previous day's area to the latter. 
# This should deal with any small discrepancies with measurements.
# first quantify the errors in the existing dataset:
pHART_growth_data[pHART_growth_data['Area_Change'] <0] 
```

```python
# create a function that will force the current item in a list to take the maximum of all of the previous items
# always ascending

def replace_descending(x):
    n_elements = len(x)
    for element_index in np.arange(n_elements):
        x.iloc[element_index] = max(x[:element_index+1])
    return x


pHART_growth_data['Area']=pHART_growth_data.groupby(['ID'])['Area'].apply(replace_descending) 
pHART_growth_data[:10]
```

```python
# reassign values for area_change
pHART_growth_data['Area_Change'] = pHART_growth_data.groupby('ID')['Area'].transform(pd.Series.diff)
```

```python
# Check there are no remaining errors in the data
pHART_growth_data ['Area_Change'] = pHART_growth_data.groupby('ID')['Area'].transform(pd.Series.diff) 
pHART_growth_data[pHART_growth_data['Area_Change'] <0]
```

```python
pHART_growth_data[:20]
```

```python
# add '% increase' column to pHART_growth_data
pHART_growth_data['% increase'] = pHART_growth_data.groupby('ID')['Area_Change'].transform(pd.Series.cumsum) 
pHART_growth_data['% increase'] = pHART_growth_data['% increase']*100
```

```python
# Add a treatment column

def define_treatment(x):
    if x[0] == 'M':
        return 'OE'
    elif x[0] == 'S':
        return 'KD'
    else:
        return 'Wt'
    
pHART_growth_data['Treatment']=pHART_growth_data['Line'].apply(define_treatment) 
pHART_growth_data[:20]
```

```python
# remove the IDs for which there is not complete data (e.g. missing days). This seems to be messing up the seaborn tsplot
g = pHART_growth_data.groupby('ID')
size = g.size()
f = size[size == 6]
only_complete = pHART_growth_data[pHART_growth_data['ID'].isin(f.index)]
```

```python
plt.suptitle('% increase for surface area of transformed and Wt liverworts ')
plt.xticks(np.arange(3, 7, 1))
lm = sns.tsplot(only_complete, time = 'Day', unit = 'ID', value = '% increase', condition = 'Line')
axes = lm.axes
axes.set_xlim(3,)
plt.legend

```

```python
# Averaged values for the treatments
plt.suptitle('% increase for surface area of transformed and Wt liverworts ')
plt.xticks(np.arange(3, 7, 1))
lm = sns.tsplot(only_complete, time = 'Day', unit = 'ID', value = '% increase', condition = 'Treatment' )
axes = lm.axes
# axes.set_xlim(3,)
```

```python
# inspect only the day 1 data to see if the starting size of the gemmae differ between treatments
starting_size = pHART_growth_data[pHART_growth_data["Day"] == 1]
x = sns.factorplot("Line","Area",data = starting_size, hue = 'Treatment', kind = 'bar', aspect = 2, hue_order = ("OE","KD","Wt"))
plt.suptitle('Gemma size on day one')
plt.xticks(rotation=-50)
```
